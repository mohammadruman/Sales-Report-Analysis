Data Analysis and also Visualize

```python
figure = px.line(data, x='Date', y='Units Sold', title='Units Sales Over Time')
#Unit sold Units Sales Over Time
```

```python
#Revenue
```

```python
figure = px.line(data, x='Date', y='Total Revenue', title='Total Revenue per Time')
```

```python
figure.show()
```

```python
category_units = data.groupby('Product Category')['Units Sold'].sum()
```

```python
category_units.plot(kind='bar', color='skyblue')
plt.title('Units Sold by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Units Sold')
plt.show()
```

```python
category_units_Revenue = data.groupby('Product Category')['Total Revenue'].sum()
```

```python
category_units_Revenue.plot(kind='bar', color='skyblue')
plt.title('Total Revenue by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Revenue')
plt.show()
```

```python
data.head()
```

```python
#Region Wise Revenue
region_revenue = data.groupby('Region')['Total Revenue'].sum()
```

```python
region_revenue = data.groupby('Region')['Total Revenue'].sum().reset_index()


fig = px.pie(
    region_revenue,
    names='Region',             # The column for labels (categories)
    values='Total Revenue',     # The column for slice sizes
    title='Total Revenue per Region'
)


fig.show()
```

```python
total_revenue = data['Total Revenue'].sum()
print("Total Revenue:", round(total_revenue))
```

```python
#Revenue Based on the product Category
revenue = data.groupby('Product Category')['Total Revenue'].sum()
pd.DataFrame(revenue)
```

```python
#revenue based on payment method
revenue_payment = data.groupby('Payment Method')['Total Revenue'].sum()
pd.DataFrame(revenue_payment)
```

```python
pd.DataFrame(revenue_payment).plot(kind='bar',title = 'Revenue based on Payment Method',ylabel = 'Revenue')
```
